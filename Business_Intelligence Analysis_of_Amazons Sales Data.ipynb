# Avg price in category

```python
#the compare_prices dataframe contains each category and the average of its products price
compare_avg_price = sorted_data.compare_prices.head(8)
print(compare_avg_price)
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x='average_price', y='category', data=sorted_data.compare_prices, orient='h')
plt.title('Boxplot of Prices by Category')
plt.xlabel('Price')
plt.ylabel('Category')
plt.show()
plt.clf()
```

# a beautiful mess!!

```python
#lets do that again but instead of every single category which amounts to 
num_cat = sorted_data.compare_prices['category'].nunique()
print('There are ', num_cat, ' categories of products. ')
```

```python
#let us only use the top 8 categories by avg price
palette = sns.color_palette('Set3')

# Create the barplot
plt.figure(figsize=(15,10))
sns.barplot(x='average_price', y='category', data=compare_avg_price, palette=palette)
plt.title('Barplot of Average Prices by Category', fontsize=20)
plt.xlabel('Average Price', fontsize=15)
plt.ylabel('Category', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

The highest average price in a category is just short of 1000 US Dollars.

# Conclusion

From the analysis, several key insights can be drawn about the dataset:

The top 10 most sold products range from Smart Watches to USB Cables, indicating a diverse range of popular products.

The highest-rated products were mostly electronic devices and accessories, showing that these items tend to satisfy customers' needs.

In terms of product pricing, it's noteworthy that the most expensive products are majorly in the electronics category such as Smart TVs and High-End Smartphones. This underlines the importance of electronics as high-value items in online sales.

Upon examining the total products in each category, it was found that USB Cables and Smart Watches were the most frequent, suggesting these are popular items among customers (propably because of ware and tear).

When we estimated revenue by category, it was found that Televisions and Smartphones generated the highest revenue. This aligns with the earlier finding of these being among the most expensive products.

A mild correlation was found between price and ratings. This indicates that more expensive items do not necessarily receive higher ratings. Quality and value for money could be a more crucial determinant of customer satisfaction than price.

Lastly, the analysis of the average price across categories revealed that air conditioners and laptops had the highest average prices.

Based on these findings, here are a few recommendations:

Inventory Management: Stock up on popular items such as USB Cables and Smart Watches to ensure availability and to increase sales.

Pricing Strategy: Since there isn't a strong correlation between price and ratings, focus on providing value for money to maintain customer satisfaction.

Marketing Strategy: Prioritize advertising for high revenue categories like Televisions and Smartphones.

Product Range: Continue to offer a diverse range of products to cater to various customer needs.

Quality Assurance: Ensure the quality of products across all price ranges to maintain high ratings and customer satisfaction.

This analysis provides valuable insights that can guide decision-making in inventory management, pricing, and marketing strategies.
